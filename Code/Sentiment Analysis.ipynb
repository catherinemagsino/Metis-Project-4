## NLP Classification

```python
df_NV_Mex_class = df_NV_Mex_restaurants[(df_NV_Mex_restaurants['stars_x']==1) | (df_NV_Mex_restaurants['stars_x']==5)]
```

```python
df_NV_Mex_class.info()
```

```python
df_NV_Mex_class['stars_x'].value_counts()
```

```python
X=df_NV_Mex_class['text']
y=df_NV_Mex_class['stars_x']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X=cv.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_sample(X_train,y_train)
```

```python
from  sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
nb.fit(X_resampled,y_resampled)
```

```python
predictions = nb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

## Using Text Processing

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
import string
from nltk.corpus import stopwords
```

```python
def text_preprocess(mess):
    """
    1. remove punc
    2. remove stop words
    3. return list of clean text words
    """
    nopunc=[char for char in mess if char not in string.punctuation]
    nopunc=''.join(nopunc)
    return[word for word in nopunc.split() if word.lower() not in stopwords.words('english')+['mexican','las','vegas']]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
from sklearn.pipeline import Pipeline
```

```python
#pipe = Pipeline([('bow',CountVectorizer(analyzer=text_preprocess)),
#                 ('tfidf',TfidfTransformer()),
#                 ('model',MultinomialNB())])

pipe = Pipeline([('bow',CountVectorizer(analyzer=text_preprocess,token_pattern="\\b[a-z][a-z]+\\b")),
                 ('model',MultinomialNB())])
```

```python
X=df_NV_Mex_class['text']
y=df_NV_Mex_class['stars_x']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train=np.array(X_train).reshape(-1, 1)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_sample(X_train,y_train)
```

```python
pipe.fit(X_resampled,y_resampled)
```

```python
predictions=pipe.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```
